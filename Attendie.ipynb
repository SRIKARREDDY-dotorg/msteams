```
#imporitng section
import pandas as pd
import numpy as np
import os
from datetime import datetime as dt
```

```
attend = pd.read_csv("/content/meetingAttendance_16_07_2020.tsv",delimiter="\t")
attend.head() 
```

```
attendance = attend['Full Name'].unique()
A = pd.DataFrame(attendance,columns=['Full Name - Roll No'])
A.head()
```

```
join = attend.loc[(attend['User Action']=="Joined")]
```

```
join.head()
```

```
newattend = attend.loc[(attend['User Action']=="Left")]
```

```
newattend.head()
```

```
dic = {}
last = '16/07/2020, 18:00:00'
for item in A['Full Name - Roll No']:
  x = attend[(attend['Full Name']==item)&(attend['User Action']=='Joined')]
  if x.shape[0]>1:
    y = attend[(attend['Full Name']==item)&(attend['User Action']=='Left')]
    for i in range(x.shape[0]-1):
      time = dt.strptime(x.iloc[i]['Timestamp'], "%d/%m/%Y, %H:%M:%S")
      time1 = dt.strptime(y.iloc[i]['Timestamp'], "%d/%m/%Y, %H:%M:%S")
      if item in dic:
        dic[item]+=(time1 - time)
        #print(dic[item])
      else:
        dic[item] = (time1-time)
        #print(dic[item])
    time = dt.strptime(x.iloc[x.shape[0]-1]['Timestamp'], "%d/%m/%Y, %H:%M:%S")
    time1 = dt.strptime(last, "%d/%m/%Y, %H:%M:%S")
    dic[item]+=(time1-time)
  else:
    time = dt.strptime(x.iloc[x.shape[0]-1]['Timestamp'], "%d/%m/%Y, %H:%M:%S")
    time1 = dt.strptime(last, "%d/%m/%Y, %H:%M:%S")
    dic[item]=(time1-time)
```

```
A['Time spent'] = dic.values()
```
